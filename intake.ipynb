```python
import numpy
from matplotlib import pyplot as plt
```

```python
a = 2
x = numpy.linspace(1, 44)

def f(t_):
    t = numpy.radians(t_)
    return numpy.sqrt(1 - numpy.sin(t) * numpy.sin(t) * a)

#print (x)
#print(f(x))
plt.plot(x, f(x))
plt.plot(x, numpy.cos(numpy.radians(x * numpy.sqrt(a))))
plt.show()
```
