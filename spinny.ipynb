```python
motor_system = MotorSystem(motor=cim, motor_count=1, gearing_ratio=1)
a = SpinSimulator(
    flywheel_radius_m=inch_to_meter(5),
    flywheel_mass_kg=lbs_to_kg(1),
    motor_system=motor_system
)
```

```python
a.run_sim(timeout=3)
vs_maker = numpy.vectorize(radps_to_rpm)
vs = vs_maker(a.vs)
plt.plot(a.ts, vs)
limit = [motor_system.free_speed] * len(a.ts)
plt.plot(a.ts, limit)
print (vs[-2])
plt.show()
```
